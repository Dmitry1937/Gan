```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image 
from pathlib import Path
import random

import numpy as np
```

```python
TRAIN_female_DIR = Path(PATH+'trainA/')
TRAIN_anime_DIR = Path(PATH+'trainB/')

SIZE = 256

train_female_files = sorted(list(TRAIN_female_DIR.rglob('*.jpg')))
train_anime_files = sorted(list(TRAIN_anime_DIR.rglob('*.jpg')))
```

```python
class Female_Anime(Dataset):
    """
      ,      
          
    """
    def __init__(self,files_anime, files_female, mode = 'Train',):
        super().__init__()
        #    
        self.files_anime = sorted(files_anime)
        self.files_female = sorted(files_female)
        #  
        self.mode = mode

        self.len_anime = len(self.files_anime)
        self.len_female = len(self.files_female)
           
    def __len__(self):
        return self.len_anime
      
    def load_sample(self, file):
        image = Image.open(file)
        image.load()
        return image
  
    def __getitem__(self, index):
        #      PyTorch   
        transform = transforms.Compose([
            transforms.Resize(int(SIZE * 1.12), ),
            transforms.RandomRotation(15,),
            transforms.RandomCrop((SIZE, SIZE)),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            transforms.Normalize(
            [0.5, 0.5, 0.5,], 
            [0.5, 0.5, 0.5,]),
        ])

        transform_anime = transforms.Compose([
          transforms.Resize(int(SIZE * 1.24), ),
          transforms.RandomRotation(35,),
          transforms.RandomCrop((SIZE, SIZE)),
          transforms.RandomHorizontalFlip(),
          transforms.ToTensor(),
          transforms.Normalize(
          [0.5, 0.5, 0.5,], 
          [0.5, 0.5, 0.5,]),
              ])
        
        transform_test = transforms.Compose([
          transforms.Resize(SIZE),
          transforms.ToTensor(),
          transforms.Normalize(
          [0.5, 0.5, 0.5,], 
          [0.5, 0.5, 0.5,]),
              ])
        
        anime = self.load_sample(self.files_anime[index])

        female = self.load_sample(self.files_female[random.randint(0, self.len_female - 1)])  
        
        if self.mode ==  'Train':   
            anime = transform_anime(anime)
            female = transform(female) 
        else:
            anime = transform_test(anime)
            female = transform_test(female) 
      
        return (anime,female)
```

```python
class Discriminator(nn.Module):
    def __init__(self,):
        super(Discriminator, self).__init__()
        nc = 3
        ndf = 64
        self.GM = None #garbage
        self.main = nn.Sequential(
 
            nn.Conv2d(nc, ndf, 4, 2, 1, bias=True),
            nn.LeakyReLU(0.2, inplace=True),
 
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias = True),
            nn.InstanceNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.MaxPool2d(2,2),
 
            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias = True),
            nn.LeakyReLU(0.2, inplace=True),
 
            nn.Conv2d(ndf * 4, ndf * 8, 4, 2, 1, bias = True),
            nn.InstanceNorm2d(ndf * 8),
            nn.LeakyReLU(0.2, inplace=True),
            nn.MaxPool2d(2,2),
 
            nn.Conv2d(ndf * 8, 1, 3, 1, 0, bias = True),
            nn.Sigmoid(),
            nn.AdaptiveAvgPool2d((1,1)),
                )
        
    def forward(self, input):
        return self.main(input).view(-1,1)
```

## Create generator

```python
def blocks(input_res = 256, num_features = 256):

    result = nn.Sequential(
        nn.Conv2d(input_res, num_features,(3,3),stride = (1,1), padding = (1,1)),
        nn.LeakyReLU(0.2, inplace=True),
        nn.Conv2d(input_res, num_features,(3,3),stride = (1,1), padding = (1,1)),
        nn.LeakyReLU(0.2, inplace=True),
        nn.Conv2d(input_res, num_features,(3,3),stride = (1,1), padding = (1,1)),
        nn.LeakyReLU(0.2, inplace=True),
        nn.InstanceNorm2d(input_res),
        )
    return result
```

```python
class Generator(torch.nn.Module):
    def __init__(self,):
        """
        In the constructor we instantiate two nn.Linear modules and assign them as
        member variables.
        """
        super().__init__()
        channels = 3

        # Initial convolution block
        out_features = 64
        #encoder
        self.input = nn.Sequential(
            nn.ReflectionPad2d(channels),
            nn.Conv2d(3, out_features,(7,7)),
            nn.InstanceNorm2d(out_features),
            nn.MaxPool2d(2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(out_features, out_features*2,(3,3), stride = (1,1), padding = (1,1)),
            nn.LeakyReLU(0.2, inplace=True),
            nn.MaxPool2d(2),
            nn.Conv2d(out_features*2, out_features*4,(3,3), stride = (1,1), padding = (1,1)),
            nn.LeakyReLU(0.2, inplace=True),
            )
        
        self.in0 = nn.InstanceNorm2d(256)
        self.block0 = blocks()
        self.block1 = blocks()
        self.block2 = blocks()
        self.block3 = blocks()
        self.block4 = blocks()
        self.block5 = blocks()
        self.block6 = blocks()
        self.block7 = blocks()
        
        self.out = nn.Sequential(
 
            nn.Upsample(scale_factor=2),
            nn.Conv2d(out_features*4, out_features*2, 3, stride=1, padding=1),
            nn.InstanceNorm2d(out_features*2),
            nn.LeakyReLU(0.2, inplace=True),
            
            nn.Upsample(scale_factor=2),
            nn.Conv2d(out_features*2, out_features, 3, stride=1, padding=1),
            nn.InstanceNorm2d(out_features),
            nn.LeakyReLU(0.2, inplace=True),
            
            
            nn.ReflectionPad2d(channels), 
            nn.Conv2d(out_features, channels, 7),
            nn.Tanh(),
        )

    def forward(self, x):
        """
        In the forward function we accept a Tensor of input data and we must return
        a Tensor of output data. We can use Modules defined in the constructor as
        well as arbitrary operators on Tensors.
        """
        x = self.input(x)
        x = self.in0(x)        
        x =   self.block0(x) + x  
        x =   self.block1(x) + x 
        x =   self.block2(x) + x 
        x =   self.block3(x) + x 
        x =   self.block4(x) + x 
        x = self.in0(x)

        out = self.out(x)

        return out
    
model_generator_female = Generator()
model_generator_female(inputs[0]).shape
```

## Train

```python
from tqdm import tqdm, tqdm_notebook
import matplotlib.pyplot as plt
from matplotlib.colors import  Normalize
import itertools
from torch.optim.lr_scheduler import MultiplicativeLR
```

```python
def fit_epoch(model_generator_female,model_classification_female, 
                model_generator_anime, model_classification_anime,
                anime_female_loader,
                criterion_generator, opt_generator,
                criterion_classification, opt_classification_female, opt_classification_anime,run
                ):
    
    running_loss_generator = 0.0
    running_loss_classification = 0.0

    processed_data_generator = 0
    processed_data_classification = 0
  
    for i, anime_female in enumerate(anime_female_loader):

        inputs_anime,inputs_female = anime_female
        batch = inputs_female.shape[0]
        real = np.ones((batch,1))
        fake = np.zeros_like(real)
        labels = np.concatenate((real,fake),axis = 0 )
 
        inputs_female = inputs_female.to(DEVICE)
        inputs_anime = inputs_anime.to(DEVICE)

        
        labels = torch.tensor(labels,dtype=torch.float32).to(DEVICE)

        kof_gan = 10
        kof_identity = 5
        kof_classification = 1
        
        #-----Train Generator anime-----------------------------------------------------#
            
        opt_generator.zero_grad()
        #loss identity
        loss_female = criterion_generator(model_generator_female(inputs_female), inputs_female)
        loss_anime = criterion_generator(model_generator_anime(inputs_anime), inputs_anime)

        #loss gan (cycle)
        outputs_female = model_generator_female (inputs_anime)
        outputs_female_anime = model_generator_anime(outputs_female.detach())
        #I take loss only for cycle Anime-Female-Anime
        loss_generator_anime_gan = criterion_generator(outputs_female_anime, inputs_anime)
        
        outputs_anime = model_generator_anime (inputs_female)
        outputs_anime_female = model_generator_female(outputs_anime.detach())
        
        #------
        # Because anime more simple than face I take loss from criterion_classification
        inputs = torch.cat((outputs_female_anime, inputs_anime), 0)
        outputs_classification_anime = model_classification_anime(inputs)
        loss_generator_anime = criterion_classification(
            outputs_classification_anime[:batch],1-labels[:batch])
        
        inputs = torch.cat((outputs_anime_female, inputs_female), 0)
        outputs_classification_female = model_classification_female(inputs)
        loss_generator_female = criterion_classification(
            outputs_classification_female[:batch],1-labels[:batch])
        #------
        
        #loss classification
        inputs = torch.cat((outputs_anime, inputs_anime), 0)
        outputs_classification_anime = model_classification_anime(inputs)
        
        loss_from_classification_anime = criterion_classification(
            outputs_classification_anime[:batch],1-labels[:batch])
        
        inputs = torch.cat((outputs_female, inputs_female), 0)
        outputs_classification_female = model_classification_female(inputs)
        loss_from_classification_female = criterion_classification(
            outputs_classification_female[:batch],1-labels[:batch])
        
        #loss all
        loss_generator = kof_classification*(loss_generator_female +loss_generator_anime)+\
            kof_classification*(loss_from_classification_female+loss_from_classification_anime)+\
            kof_identity*(loss_female+loss_anime)+loss_generator_anime_gan*kof_gan

        loss_generator.backward()
        opt_generator.step()
        
        
       #-----Train Discriminator anime-----------------------------------------------------#

        opt_classification_anime.zero_grad()

        outputs_anime = model_generator_anime (inputs_female).detach()
        #outputs_anime_female = model_generator_female(outputs_anime).detach()

        inputs = torch.cat((outputs_anime, inputs_anime), 0)
        outputs = model_classification_anime(inputs)

        loss_classification = criterion_classification(outputs,labels,)
        loss_classification.backward()    

        opt_classification_anime.step()

        
       #-----Train Discriminator female-----------------------------------------------------#

        opt_classification_female.zero_grad()

        outputs_female = model_generator_female (inputs_anime).detach()
        #outputs_female_anime = model_generator_anime(outputs_female).detach()

        inputs = torch.cat((outputs_female, inputs_female), 0)
        outputs = model_classification_female(inputs)

        loss_classification = criterion_classification(outputs,labels,)
        loss_classification.backward()    

        opt_classification_female.step()
        
        
        #----------------------------------------------------------------------------------#
        
        running_loss_generator += loss_generator.item() * batch
        running_loss_classification += loss_classification.item() * inputs.shape[0]

        processed_data_generator += batch
        processed_data_classification += inputs.size(0)
                
        train_loss_generator = running_loss_generator / processed_data_generator
        train_loss_classification = running_loss_classification / processed_data_classification
      
        if (i%100) == 99 :

            IPython.display.clear_output(wait=True)
            print(i,train_loss_generator, train_loss_classification, sep = '\n' )
            plt.figure(figsize=(18, 6))
            result = torch.cat((outputs_anime, inputs_female), 0).detach().cpu()
            k_max = result.size(0)//2
          
            if k_max>3:
                k_max=3
            for k in range(k_max):
                plt.subplot(2, 3, k+1)
                plt.imshow(np.rollaxis(result[k].numpy()/2+0.5, 0, 3), )
                plt.title('Example')
                #plt.axis('off')
              
                plt.subplot(2, 3, k+4)
                plt.imshow(np.rollaxis(result[result.size(0)//2+k].numpy()/2+0.5, 0, 3), )
                plt.title('Real')
                #plt.axis('off')

        plt.show()

    return  train_loss_generator, train_loss_classification, torch.cat((outputs_anime, inputs_female), 0)
```

```python
#last lr
LR = 0.00005*0.9**18
LR
```

```python
def train(anime_female_dataset, model_generator_female, model_classification_female,\
          model_generator_anime, model_classification_anime, epochs, batch_size, run = False):
    
    anime_female_loader = DataLoader(anime_female_dataset, batch_size=batch_size, shuffle=True)

    history = []
    log_template = "\nEpoch {ep:03d} train_loss_generator_anime: {t_loss:0.4f} train_loss_classification_anime: {t_loss2:0.4f}"
    
    with tqdm(desc="epoch", total=epochs) as pbar_outer:

        criterion_classification = nn.BCELoss()

        opt_classification_female = torch.optim.Adam(model_classification_female.parameters(),lr = LR, betas = (0.5, 0.999))
        opt_classification_anime = torch.optim.Adam(model_classification_anime.parameters(),lr = LR, betas = (0.5, 0.999))
        scheduler1 = MultiplicativeLR (opt_classification_female, lr_lambda = lambda epoch: 0.9)
        scheduler2 = MultiplicativeLR (opt_classification_anime, lr_lambda = lambda epoch: 0.9)  


        criterion_generator = nn.L1Loss()   

        opt_generator = torch.optim.Adam(
            itertools.chain(model_generator_female.parameters(), model_generator_anime.parameters()),\
            lr = LR, betas = (0.5, 0.999))
        scheduler3 = MultiplicativeLR (opt_generator, lr_lambda = lambda epoch: 0.9)  
        
        for epoch in range(epochs):

            train_loss_generator, train_loss_classification, outputs_anime = fit_epoch(
                                            model_generator_female,model_classification_female, 
                                            model_generator_anime, model_classification_anime,
                                            anime_female_loader,
                                            criterion_generator, opt_generator,
                                            criterion_classification, opt_classification_female, opt_classification_anime,run
                                            )
                
            scheduler1.step()
            scheduler2.step()
            scheduler3.step()

            history.append((train_loss_generator, train_loss_classification ))
            
            pbar_outer.update(1)

            # Visualize tools
            IPython.display.clear_output(wait=True)
            tqdm.write(log_template.format(ep=epoch+1, t_loss=train_loss_generator, t_loss2 = train_loss_classification ))
            
            # show example
            result = outputs_anime.detach().cpu()# detach and put into cpu
            plt.figure(figsize=(18, 6))
            k_max = result.size(0)//2
            if k_max>3:
                k_max=3
            for k in range(k_max):
                plt.subplot(2, 3, k+1)
                plt.imshow(np.rollaxis(result[k].numpy()/2+0.5, 0, 3), )
                plt.title('Example')
                #plt.axis('off')
                
                plt.subplot(2, 3, k+4)
                plt.imshow(np.rollaxis(result[result.size(0)//2+k].numpy()/2+0.5, 0, 3), )
                plt.title('Real')
                #plt.axis('off')

            plt.show()

            torch.save(model_generator_female.state_dict(), f'/content/drive/My Drive/Gan/save/model_generator_female{epoch}.pt')
            torch.save(model_generator_anime.state_dict(), f'/content/drive/My Drive/Gan/save/model_generator_anime{epoch}.pt')
            torch.save(model_classification_female.state_dict(), f'/content/drive/My Drive/Gan/save/model_classification_female{epoch}.pt')
            torch.save(model_classification_anime.state_dict(), f'/content/drive/My Drive/Gan/save/model_classification_anime{epoch}.pt')
       
    return result, history
```

```python
result, history = train(anime_female_dataset, model_generator_female, model_classification_female,\
                model_generator_anime, model_classification_anime, EPOCHS, BATCH_SIZE, run = True)
```
